```python
clean_df = df.copy() # Create a copy of the DataFrame to work with
```

```python
clean_df.head(3) # Display the first 3 rows of the cleaned DataFrame
```

```python
df['Fare'] # Display the 'Fare' column of the DataFrame
```

```python
# Repare the 'Age' column by filling missing values with the median
median_age = clean_df['Age'].median() # calculate the median age
clean_df['Age'] = clean_df['Age'].fillna(median_age) # fill missing values with the median age
```

```python
clean_df['Age'].isnull().sum() # Check if there are any missing values left in the 'Age' column
```

```python
# Repare the 'Embarked' column by filling missing values with the mode
embarked_mode = clean_df['Embarked'].mode()[0]  # calculate the mode of the 'Embarked' column
clean_df['Embarked'] = clean_df['Embarked'].fillna(embarked_mode)  # fill missing values with the mode
```

```python
clean_df['Embarked'].isnull().sum()  # Check if there are any missing values left in the 'Embarked' column
```

```python
# Create new caracteristics based on existing ones
clean_df['Has_Cabin'] = clean_df['Cabin'].notnull().astype(int)  # Create a new column indicating if the passenger has a cabin
clean_df['Cabin'] = clean_df['Cabin'].fillna('Unknown')  # Fill missing cabin values with 'Unknown'
```
